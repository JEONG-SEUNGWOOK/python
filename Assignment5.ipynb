#### 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

  - count: Counter 인스턴스가 지니는 초기 정수 값
  - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

   - print c 
   - [Count (step: 1)] 11 
   - print d 
   - [Count (step: 2)] 12 

- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.
   - c() 
   - d() 
   - print c 
   - [Count (step: 1)] 12 
   - print d 
   - [Count (step: 2)] 14 

- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
   - c = c + 5 
   - d = d - 5 
   - print c 
   - [Count (step: 1)] 17 
   - print d 
   - [Count (step: 2)] 9 

- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

   - print c > 10 
   - True 
   - print d > 10 
   - False 
   - print c < 10 
   - False 
   - print d < 10 
   - True 
   - print c == 17 
   - True 
   - print d != 9 
   - False 

```python
class Count:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count(Step:%d)] %d" %(self.step, self.count)
    def incr(self):
        self.count = self.count + self.step
        return self.count
    def __call__(self):
        self.count = self.count + self.step
        return self.count
    def __add__(self, other):
        self.count = self.count + other
        return self.count
    def __sub__(self, other):
        self.count = self.count - other
        return self.count
    
    def __lt__(self, other):
        return self.count > other
    def __gt__(self, other):
        return self.count < other
    def __eq__(self, other):
        return self.count == other
    def __ne__(self, other):
        return self.count != other
```

```python
c = Count(10)
d = Count(10, 2)

print c
print d
print

c.incr()
d.incr()
print c
print d
print

c()
d()

print c
print d
print 

c = c + 5
d = d - 5
print c
print d
print

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

#### 설명

먼저 \__init\__은 해당 클래스의 생성자를 정의하는 함수로써 l이라는 객체를 받아 각 인스턴트 객체에 추가한다. 그리고 eliminate_duplicate라는 메소드를 실행하여 중복값을 없앤다

eliminate_duplicate라는 메소드는 정적인 메소드이며 리스트 객체를 받아와 s라는 리스트에 추가하되 s안에 이미 같은 객체가 있으면 추가하지 않는다. 그리고 다시 리스트 s를 리스트 l에 담는다.

\__str\__메소드는 MySet: {...}라는 형식으로 출력하기 위한 메소드이다.

이제 메인을보면 이 함수가 메인이라는 모듈에서 불릴때만 조건문 안의 내용이 실행된다. 먼저 s에 1, 2, 2, 3이 리스트로 들어가고 출력을 하면 위 과정을 거쳐 중복이 제거되어 커스텀한 형식으로 출력하게된다.

#### [5번 문제]
#### 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

#### 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

 - u = s | t 
 - print u 
  - MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

 - u = s & t 
 - print u 
  - MySet: {2, 3} 

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

 - s = MySet([1, 2, 3]) 
 - t = MySet([3, 4, 5]) 
 - u = s - t 
 - print u 
  - MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l): 
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other): #두 리스트를 받아 l이라는 리스트에 입력 후 eliminate_duplicate()메소드를 이용해 중복값을 제거후 정렬한뒤 str에 정의된 형식으로 출력
        l = []
        l.extend(self)
        l.extend(other)
        MySet.eliminate_duplicate(l)
        l.sort()
        return MySet(l)
    
    def __and__(self, other): #other리스트 내의 원소가 self리스트 내에 같은 원소가 있으면 result리스트에 추가한 후 정렬, 출력
        result = []
        for l in other:
            if l in self:
                result.append(l)
        result.sort()
        return MySet(result)
    
    def __sub__(self, other): #두 리스트의 교집합을 li에 저장 후 self를 담는 result내에 li의 원소와 중복되는 원소를 제거 후 정렬, 출력
        li = self & other
        result = self
        for l in li:
            if l in self:
                result.remove(l)
        result.sort()
        return MySet(result)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
    u = s | t
    print u
    
    u = s & t
    print u
    
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print u
```

1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고, 각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다.

1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.
  
  예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자,
  
  115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.

```python
num_dic = {1:'one', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six', 7:'seven', 8:'eight'
, 9:'nine', 10:'ten', 11:'eveven', 12:'twelve', 13:'thirteen', 14:'fourteen', 15:'fifteen'
, 16:'sixteen', 17:'seventeen', 18:'eighteen', 19:'nineteen', 20:'twenty', 30:'thirty', 40:'forty'
, 50:'fifty', 60:'sixty', 70:'seventy', 80:'eighty', 90:'ninety', 100:'hundred', 1000:'thousand'}

def StringToNumber(num):
    if num == 0 :
        return 0
    
    if num > 0 and num <= 20:
        return len(num_dic[num])
    
    if num > 20 and num < 100:
        unit_digit = num%10
        tens_digit = num-unit_digit
        if unit_digit == 0:
            return 0 + len(num_dic[tens_digit])
        else :    
            return len(num_dic[unit_digit]) + len(num_dic[tens_digit])
    
    if num >= 100 and num < 1000:
        hundred_digit = num/100
        reminder = num%100
        reminder = StringToNumber(reminder)
        if reminder == 0:
            return len(num_dic[hundred_digit]) + len(num_dic[100]) + 0
        else :
            return len(num_dic[hundred_digit]) + len(num_dic[100]) + len('and') + reminder
            
    
    if num == 1000:
        return len(num_dic[1]) + len(num_dic[1000])
    
    else:
        return '1~1000의 숫자만 입력해주세요'
    
sum_of_num = 0
for x in range(1, 1001):
    sum_of_num += StringToNumber(x)

print sum_of_num
```

#### 설명
1부터 20까지의 숫자는 고유한 지칭 단어가 있으므로 각각 사전으로 정의하고, 각 10단위와 100, 1000을 정의한다.

입력된 숫자가 1~20의 숫자이면 그 숫자의 키에 해당되는 값의 길이를 반환하고, 21~99의 숫자는 10의 자리수와 1의 자리수를 분리하여 10의자리수의 길이를 구하고 1의자리수 또한 길이를 구하지만, 만약 1의자리수가 0인경우 길이 0을 반환한다. 100에서 999의 숫자는 100의자리수를 숫자를 구하여 해당 숫자 + 100의 길이와 100의자리수를 제외한 나머지 숫자를 재귀적으로 구하여 길이+ and의 길이를 반환해준다. 역시 마찬가지로 나머지숫자가 0인경우 길이 0을 반환.

1000이 입력되면 one-hundred의 길이가 반환되고 1~1000이외의 숫자가 입력될 경우 '1~1000'의 숫자만 입력해주세요 라는 문구가 출력

```python
tri = '''75
95 64
17 47 82
18 35 87 10
20 04 82 47 65
19 01 23 75 03 34
88 02 77 73 07 63 67
99 65 04 28 06 16 70 92
41 41 26 56 83 40 80 70 33
41 48 72 33 47 32 37 16 94 29
53 71 44 65 25 43 91 52 97 51 14
70 11 33 28 77 73 17 78 39 68 17 57
91 71 52 38 17 14 91 43 58 50 27 29 48
63 66 04 68 89 53 67 30 73 16 69 87 40 31
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23'''

li = [[int(e) for e in line.split()] for line in tri.split('\n')][::-1]

l = len(li)
for i in range(l):
    for j in range(l - i-1):
        li[i+1][j] += max(li[i][j],li[i][j+1])

print li[-1][-1]
```

1900년 1월 1일은 월요일이다.

4월, 6월, 9월, 11월은 30일까지 있고, 1월, 3월, 5월, 7월, 8월, 10월, 12월은 31일까지 있다.

2월은 28일이지만, 윤년에는 29일까지 있다.

윤년은 연도를 4로 나누어 떨어지는 해를 말한다. 하지만 400으로 나누어 떨어지지 않는 매 100년째는 윤년이 아니며, 400으로 나누어 떨어지면 윤년이다

20세기 (1901년 1월 1일 ~ 2000년 12월 31일) 에서, 매월 1일이 일요일인 경우는 총 몇 번입니까?

```python
year = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
leap_year = {1:31, 2:29, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

def isLeapYear(year):
    if (year % 4 == 0) or (year % 400 == 0 and year % 100 != 0):
        return True
    else :
        return False

def whatDayIsNewYear(year):
    day = 1
    for x in range(1900, year):
        if isLeapYear(x) == True:
            day += 366
        else:
            day += 365
            
    return day % 7 # 0:일요일  1:월요일  2:화요일  3:수요일  4:목요일  5:금요일  6:토요일

def howManySunday(from_year, to_year):
    day = whatDayIsNewYear(from_year)
    if day == 0:
        cnt_of_sunday = 1
    else :
        cnt_of_sunday = 0
        
    for x in range(from_year, to_year+1):
        if isLeapYear(x) == True:
            for y in range(1, 13):
                day += leap_year[y]
                if day%7 == 0:
                    cnt_of_sunday += 1
        else :
            for y in range(1, 13):
                day += year[y]
                if day%7 == 0:
                    cnt_of_sunday += 1
    return cnt_of_sunday

print howManySunday(1901, 2000)
```

n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때, 서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다.

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.

또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.

따라서 220과 284는 친화쌍이 됩니다.

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.

```python
def d(num):
    li = []
    l2 = []
    sqrt_n = int(num**0.5)+1
    for x in range(1, sqrt_n):
        if num % x == 0 :
            li.append(x)
    
    for l in li:
         l2.append(l)
    
    for l in l2:
         li.append(num/l)
            
    sum_l = 0
    for l in li:
        if l != num:
            sum_l += l
    return sum_l

num = 10000
li_s = 0
for x in range(1, num):
    a = d(x)
    b = d(a)

    if x == b and x != a:
        
        li_s += x

print li_s
```

```python
# -*-coding:utf-8 -*-
import string
import urllib2
import pickle
from string import maketrans


pun = string.punctuation
stopwords = [ 'A', 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', \
             'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', \
             'or', 's', 'such', 't', 'that', 'the', 'their', 'then', \
             'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
  

def read_url(url):
    source = urllib2.urlopen(url).read()
    return source

def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:],"")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:],"")
    
    return string

#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:],"")
    return string

#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:],"")
    return string

#구두문자 제거 함수
def punc_delete(source):
    for i in range (len(pun)):
        source= source.replace(pun[i]," ")
        
    return source
    
    
def cntWords(s):
    #공백 단위로 단어 쪼개고 단어 넣을 딕셔너리 선언
    listSplited = s.split()
    dicWords = {}

    #단어 쪼개놓은 리스트에서 각각 단어의 개수 세기
    for i in listSplited:
        if dicWords.has_key(i)  is False:
            dicWords[i]=1
        else:
            dicWords[i] += 1
    
    return dicWords

def stopword_delete(source):
    for s in stopwords:
        if s in source: 
            del source[s]
    return source
        
def extract(url):
    source = read_url(url)
    source = tag_delete(source)
    source = javascript_delete(source)
    source = stylesheet_delete(source)
    source = punc_delete(source)
    source = cntWords(source)
    source = stopword_delete(source)
    return source

class WebWordsFrequency:
    def __init__(self, *args):
        self.url = []
        for li in args:
            self.url.append(li)
            
    def addUrl(self, newUrl):
        self.url.append(newUrl)
        return self.url
    
    def removeUrl(self, rmUrl):
        if rmUrl in self.url:
            self.url.remove(rmUrl)
        return self.url
    def listUrls(self):
        for li in self.url:
            print li
            
    def getWordsFrequency(self):
        for li in self.url:
            print extract(li)
        
    def getMaxFreqencyWords(self):
        def val(wDic):
            return wDic[1]
        
        maxFreWord = []
        for li in self.url:
            l = extract(li)
            d1 = l.keys()
            d2 = l.values()
            a = zip(d1, d2)
            a.sort(cmp=None, key=val, reverse=True)
            maxFreWord.append(a[0])
        maxFreWord.sort(cmp=None, key=val, reverse=True)
        print maxFreWord[0][0]
        for li in maxFreWord:
            if li[1] == maxFreWord[0][1] and li[0] != maxFreWord[0][0]:
                print li[0]
```
