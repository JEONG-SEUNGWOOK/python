####range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
- 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

 - range(stop)
   - range(start, stop[, step])


- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 
- 3) frange 사용 예
 - 3-1) frange(0.5)
   - [0.0, 0.1, 0.2, 0.3, 0.4]

 - 3-2) frange(1.0, 2.0)
   - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]

 - 3-3) frange(2.2, 4.0, 0.5)
   - [2.2, 2.7, 3.2, 3.7]




```python
import math

def frange(stop, start = None, step = 0.1):
    if start == None:
        start = 0.0
    else : 
        temp = start
        start = stop
        stop = temp
    i = start
    l = []
    while i < stop:
        l.append(round(i, 10))
        i += step
    return l
print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0 ,0.5)
print frange(100.0, 110.0, 1.1)
```

인자가 하나면 stop의 인자가되고 두개면 start, stop의 인자가 된다. start의 기본값을 0으로 주기위해서는 기본값이 없는 stop이 start보다 먼저와야하는데 인자를 두개를 넣을 경우 서로 뒤바뀌는 문제가 발생하기때문에 start의 기본값을 None으로 주고 인자가 하나일때는 start =0.0으로 주고 인자가 두개일 경우에는 start와 stop의 값을 바꾸도록 했다. 그리고 start부터 리스트에 추가한 후 step만큼 더하고 stop값을 넘지 않을 때까지 리스트에 계속 추가하는 방식으로 작동함.

- 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
  - split
  - map
  - join
- 2) myinitial() 함수 사용 예
  - myinitial("as soon as possible")
  - ASAP

```python
def up(x):
    return x.upper()[0]

def myinitial(word_li):
    word_li = word_li.split(' ')
    word_li = map(up, word_li)
    state = ''.join(word_li)
    return state

print myinitial('as soon as possible')
```

먼저 split를 이용하여 문장의 각 단어들을 리스트화 시킨다. 그리고 단어를 대문자로 바꾸고 각 단어의 첫글자만 가져오는 up함수를 정의한 후 map 내장함수를 사용하여 각 단어의 첫글자만 가져온후 join함수를 이용하여 하나의 단어로 합친다.

####음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if n <= 0 :
        return 0
    elif n == 1:
        return 1
    else :
        return n*myfact(n-1)
n = input("input number : ")
print myfact(n)
```

0이하의 정수를 입력하면 0을 반환하고, 1일때는 1을 반환하고 1이상의 정수일 경우에는 자기자신 n과 factorial(n-1)을 곱해준 값을 반환하여 재귀적으로 작동하게함

####양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

n → n / 2 (n이 짝수일 때)

n → 3 n + 1 (n이 홀수일 때)

####13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
####아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
####(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

####그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

####참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
def hailstone(num):
    max_cnt = 0
    max_n = 0
    
    for x in range(1, num): 
        n = x
        cnt = 1
        while True:
            if n%2 == 0:
                n = n/2
                cnt += 1
            elif n%2 == 1:
                n = 3*n + 1
                cnt += 1
            if n == 1 :
                if max_cnt<cnt :
                    max_cnt = cnt
                    max_n = x
                    break
                else:
                    break
        
    print max_n
    print max_cnt

hailstone(1000000)
```

1부터 백만까지의 숫자들을 위 계산을 연산하고 연산할 때마다 카운트를 해준다. 연산이 끝나면 max_cnt와 비교하여 가장 많이 연산한 수를 찾아낸다.

####그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?

```python
def fac(a):
    if a == 1:
        return 1
    else:
        return a*fac(a-1)

def min_path(n,m):
    return fac(n+m)/(fac(n)*fac(m))

print min_path(20,20)
```

####5) 위 문제에서 저장한 모든 words.frequency 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들에서 가장 많이 출현한 단어 3개를 뽑아 제시하시오.
 - 반드시 pickle 모듈로 저장한 5개 이상의 words.frequency를 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함

```python
# -*-coding:utf-8 -*-
import string
import urllib2
import pickle
from string import maketrans


pun = string.punctuation
stopwords = [ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', \
             'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', \
             'or', 's', 'such', 't', 'that', 'the', 'their', 'then', \
             'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
  
url = ['http://www.cnn.com', 'https://www.whitehouse.gov/',\
       'http://www.apple.com/', 'https://parse.com/', \
       'http://www.ucr.edu/', 'http://www.santamonica.com/']


def read_url(url):
    source = urllib2.urlopen(url).read()
    return source

def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:],"")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:],"")
    
    return string

#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:],"")
    return string

#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:],"")
    return string

#구두문자 제거 함수
def punc_delete(source):
    outtab = " " * 32
    trantab = maketrans(pun, outtab)
    source = source.translate(trantab)
        
    return source
    
    
def cntWords(s):
    #공백 단위로 단어 쪼개고 단어 넣을 딕셔너리 선언
    listSplited = s.split()
    dicWords = {}

    #단어 쪼개놓은 리스트에서 각각 단어의 개수 세기
    for i in listSplited:
        if dicWords.has_key(i)  is False:
            dicWords[i]=1
        else:
            dicWords[i] += 1
    

    return dicWords
        
def file_name(url):
    if url[:7] == 'http://':
        url = url[7:]
    if url[:8] == 'https://':
        url = url[8:]
    if url[-1] == '/':
        url = url[:len(url)-1]
    return url

def html_write(url,source):
    url = file_name(url)
    f = open(url+'.html','w')
    f.write(source)
    f.close()
    
def words_fre_write(url, source):
    url = file_name(url)
    f = open(url+'.words.frequency','w')
    words = cntWords(source)
    
    for s in stopwords:
        if(s in words): 
            del words[s]
            
    pickle.dump(words, f)
    f.close()
    
def words_fre_read(url):
    url = file_name(url)
    f = open(url+'.words.frequency','r')
    x = pickle.load(f)
    f.close()
    return x

def val(t):
    return t[1]

def extract(url):
    source = read_url(url)
    html_write(url,source)
    source = tag_delete(source)
    source = javascript_delete(source)
    source = stylesheet_delete(source)
    source = punc_delete(source)
    words_fre_write(url, source)
    a = words_fre_read(url)
    return a
```

```python
a = {}
for i in range(len(url)):
    a.update(extract(url[i]))
d1 = a.keys()
d2 = a.values()
a2 = zip(d1, d2)
a2.sort(cmp=None, key=val, reverse=True)
print a2[0], a2[1], a2[2]
```

처음에 url을 받아와 source에 담아주고 해당 url 태그를 html_write()를 통해서 url.html 파일을 만들어 저장한다. 그리고 source의 태그, javascript, css, 구두문자를 지우고 words_fre_write()에서 영어 불용문을 지우고 url.words.frequancy()를 통해 모든 필요없는 문자를 제거하고 단어 별 빈도수쌍의 사전을 저장한다. 저장한 값을 다시 불러와 a={}에 저장하고 사전의 빈도수값을 내림차순으로 sort해주고, 가장 많은값 3개 [0], [1], [2]를 프린트해준다.
