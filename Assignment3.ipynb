####1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addall(list):
    sum = 0
    for i in range(0, len(list)):
        sum += list[i]
    print sum
    
addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def addall(list):
    print sum([x for x in list])

addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def addallodd(list):
    sum_odd = 0
    for i in range(0, len(list)):
        if(list[i] % 2 == 1):
            sum_odd += list[i]
    print sum_odd
addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])    
```

####이 때 10,001번째의 소수를 구하세요.

```python
num = 2
cnt = 0

def primes(num):
    for x in range(2, int(num**0.5)+1):
        if num % x == 0:
            return False
    return True

while True:
    if primes(num) == True:
        cnt = cnt + 1
        if cnt == 10001:
            print num
            break
            
    num = num + 1
```

```python
# -*-coding:utf-8 -*-
import string
pun = string.punctuation
print pun
print len(pun)
```

설명 : 구두문자 목록

```python
import urllib2
from string import maketrans  

source = urllib2.urlopen("http://www.naver.com").read()
source2 = urllib2.urlopen("http://www.daum.net").read()
source3 = urllib2.urlopen("http://www.nytimes.com").read()
```

설명 : 과제에 제시된 url의 소스를 각 변수에 담는다

```python
def trans(source):
    outtab = " " * 32
    trantab = maketrans(pun, outtab)
    source = source.translate(trantab)
```

설명 : 구두문자 32개와 각 공백 32개를 maketrans함수를 통해 일대일대응으로 공백으로 치환

```python
def cntWords(s):
    #공백 단위로 단어 쪼개고 단어 넣을 딕셔너리 선언
    listSplited = source.split()
    dicWords = {}

    #단어 쪼개놓은 리스트에서 각각 단어의 개수 세기
    for i in listSplited:
        if not i in dicWords.keys():
            dicWords[i]=1
        else:
            dicWords[i]=dicWords[i]+1

    #포맷을 이용한 출력
    format = "%s : %d"

    for i in dicWords:
        print format % (i,dicWords[i])
```

설명 : 사전화 시켜주는 함수

아래는 결과물 출력

```python
trans(source)
cntWords(source)
```

```python
trans(source2)
cntWords(source2)
```

```python
trans(source3)
cntWords(source3)
```
